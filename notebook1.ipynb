# Win Rate

```python
df_merged['is_win'] = df_merged['Closed PnL'] > 0
win_rate_df = df_merged.groupby('Account')['is_win'].mean() * 100
win_rate_df
```

```python
direction_map = {
    'Open Long': 'Long', 'Close Long': 'Long', 'Long > Short': 'Long', 'Buy': 'Long',
    'Open Short': 'Short', 'Close Short': 'Short', 'Short > Long': 'Short', 'Sell': 'Short',
    'Liquidated Isolated Short': 'Short'}
df_merged['Trade_Direction'] = df_merged['Direction'].map(direction_map)


df_merged['is_win'] = df_merged['Closed PnL'] > 0
win_rate = df_merged.groupby('Account')['is_win'].mean() * 100


df_merged['Leverage'] = df_merged['Size USD'] / df_merged['Start Position']


df_merged['Leverage'] = df_merged['Leverage'].replace([float('inf'), -float('inf')], 0).fillna(0)

leverage_stats = df_merged.groupby('Account')['Leverage'].agg(['mean', 'median', 'std']).rename(
    columns={'mean': 'Avg_Leverage', 'median': 'Median_Leverage', 'std': 'Leverage_Volatility'})


ls_counts = df_merged.groupby(['Account', 'Trade_Direction']).size().unstack(fill_value=0)
ls_counts['LS_Ratio'] = ls_counts.get('Long', 0) / (ls_counts.get('Short', 0) + 1e-6)

trader_analysis = pd.DataFrame({
    'Win_Rate_Pct': win_rate
}).join([leverage_stats, ls_counts['LS_Ratio']]).reset_index()

print("Trader Performance Analysis:")
display(trader_analysis)
```

```python
from IPython.utils import process
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, classification_report
from sklearn.ensemble import RandomForestClassifier

class Trade_predictor_pipeline:
  def __init__(self):
    self.model =  RandomForestClassifier(n_estimators=100,max_depth=7, class_weight='balanced', random_state=42)
    self.scaler = StandardScaler()
    self.is_trained = False


  def preprocess_data(self,df):
    df_merged['is_win'] = (df_merged['Closed PnL'] > 0).astype(int)
    df_merged['Leverage'] = (df_merged['Size USD']/(df_merged['Start Position']+1e-6)).abs().clip(upper=125)

    daily = df_merged.groupby(['Account','date']).agg({
        'Size USD':'mean',
        'Leverage':'mean',
        'Order ID':'count',
        'Closed PnL':'sum'
    }).reset_index()

    sentiment_dummies = pd.get_dummies(df.set_index(['Account', 'date'])['classification']).groupby(level=[0,1]).first()
    daily = daily.merge(sentiment_dummies, left_on=['Account', 'date'], right_index=True)

    daily = daily.sort_values(['Account', 'date'])
    daily['target'] = (daily.groupby('Account')['Closed PnL'].shift(-1) > 0).astype(int)
    return daily.dropna()

  def train(self,df_merged):
    processed_df = self.preprocess_data(df_merged)
    features = [col for col in processed_df.columns if col not in ['Account','date','Closed PnL','target']]
    x = processed_df[features]
    y = processed_df['target']
    x_scaled = self.scaler.fit_transform(x)
    self.model.fit(x_scaled,y)
    self.is_trained = True
    return features

  def get_trader_ranking(self,df_merged):
    processed_df = self.preprocess_data(df_merged)
    features = [col for col in processed_df.columns if col not in ['Account','date','Closed PnL','target']]
    x_scaled = self.scaler.transform(processed_df[features])
    processed_df['Success_Probability'] = self.model.predict_proba(x_scaled)[:, 1]
    rankings = processed_df.groupby('Account')['Success_Probability'].mean().sort_values(ascending=False)
    return rankings

pipeline = Trade_predictor_pipeline()
feature_names = pipeline.train(df_merged)
top_traders = pipeline.get_trader_ranking(df_merged)

print("\n Top 5 Ranked Traders (Predicted Success)")
print(top_traders.head(5))
```

```python
from sklearn.model_selection import TimeSeriesSplit
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Use TimeSeriesSplit instead of random shuffle
# This respects the chronological order of trades
tscv = TimeSeriesSplit(n_splits=5)

processed_df = pipeline.preprocess_data(df_merged)
features = [col for col in processed_df.columns if col not in ['Account', 'date', 'Closed PnL', 'target']]
X = processed_df[features].values
y = processed_df['target'].values

# 2. Run the Walk-Forward Validation
for train_index, test_index in tscv.split(X):
    X_train, X_test = X[train_index], X[test_index]
    y_train, y_test = y[train_index], y[test_index]

    # Re-scale based ONLY on training data to prevent leakage
    X_train_scaled = pipeline.scaler.fit_transform(X_train)
    X_test_scaled = pipeline.scaler.transform(X_test)

    pipeline.model.fit(X_train_scaled, y_train)

# 3. Final Prediction & Metrics
y_pred = pipeline.model.predict(X_test_scaled)

print(f"--- Realistic Model Accuracy: {accuracy_score(y_test, y_pred):.2%}")
print("\n--- Final Classification Report ---")
print(classification_report(y_test, y_pred))
```
